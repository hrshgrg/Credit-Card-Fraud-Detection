```python
card.drop_duplicates(subset=None, inplace=True)
card.shape;
sns.countplot(x= 'Class',data=card).set_yscale('log');
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
features=card[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V20','V21', 'V22', 'V23', 'V25', 'V26', 'V27','Class']]
x=features.iloc[:,:-1]
y=features.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
y_pred=log_reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
conf_matrix=pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['True:0','True:1'])
plt.figure(figsize = (8,5))
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu");
```

```python
y_pred_prob=log_reg.predict_proba(x_test)
y_pred_prob_df=pd.DataFrame(data=y_pred_prob, columns=['Prob of Not Fraud','Prob of Fraud'])
y_pred_prob_df
```
